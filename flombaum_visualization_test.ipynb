Flombaum data (Ocean Atlas formatted)  
historical abundance of Prochlorococcus and Synechococcus from 1987 to 2008 from various studies, compiled by Flombaum et al. PNAS 2013.
There are 47500 records, great complement to SeaFlow data!



```python
import numpy as np
import pandas as pd
import folium



def load_csv(fname):
    df = pd.read_csv(fname, sep=',', date_parser = pd.to_datetime)
    return df


df = load_csv('flombaum_template.csv')

"""only taking depth above 5m"""
df = df[df['depth'] <= 5]
"""folium can't render many points"""
df = df.head(500)


folium_map = folium.Map(location=[40.738, -73.98],
                        zoom_start=2,
                        tiles="CartoDB dark_matter")



def add_points_to_map(name, scaling,color):
    for row in df.itertuples(index=True, name='Pandas'):
        
        popup_text = """depth (m): {} <br>
        total prochlor: {} <br>
        total synecho: {}"""
        
        popup_text = popup_text.format(
                               getattr(row, "depth"),
                               getattr(row, "prochloro_abundance"),
                               getattr(row, "synecho_abundance"))     
        
        try:
            folium.CircleMarker(location=(float(getattr(row, "lat")),
                                          float(getattr(row, "lon"))),
                                radius=(getattr(row, name)/scaling), color=color, fill=True,
                                popup = popup_text
                               ).add_to(folium_map)
        except:
            pass
        

# add prochloro -- prochloro_abundance
add_points_to_map("prochloro_abundance", 8000.0, "#0A8A9F")
add_points_to_map("synecho_abundance", 8000.0, "#E37222")



folium_map


```
